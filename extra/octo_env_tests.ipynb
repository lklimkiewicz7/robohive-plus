# Joints

```python
import gym 
import robohive
import matplotlib.pyplot as plt
```

```python
env = gym.make('Ur5Cube-OctoJoints-v0')
start_obs, info = env.reset()
```

```python
plt.imshow(start_obs['image_primary'])
start_obs['proprio']
```

```python
for i in range(1000):
    obs, reward, done, info = env.step(start_obs['proprio']);
```

```python
plt.imshow(obs['image_primary'])
obs['proprio']
```
